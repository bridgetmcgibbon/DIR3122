```python
import numpy as np
from matplotlib import pyplot
%matplotlib inline

#defining a new function linearconv w/ argument nx
#this is Step 1 (the linear convection equation), with the argument nx being the number of points in the 'grid'
def linearconv(nx):
    dx = 2/(nx-1)
    print("dx = ", dx, '\n')
    nt = 20 
    dt = 0.025
    c = 1
    
    u = np.ones(nx)
    u[int(0.5/dx):int(1/dx)+1] = 2
    
    un = np.ones
    
    for n in range(nt):
        un = u.copy()
        for i in range(1, nx):
            u[i] = un[i] - c*dt/dx*(un[i]-un[i-1])

    pyplot.plot(np.linspace(0,2,nx), u)
```
