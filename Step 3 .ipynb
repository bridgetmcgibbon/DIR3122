```python
import numpy as np
from matplotlib import pyplot
%matplotlib inline
```

```python
nx = 81
dx = 2/(nx-1)
nt = 20
nu = 0.3 #viscosity

sigma = 0.2 #sigma is some parameter, to be explained late
dt = sigma * dx**2 / nu #dt is defined by sigma

u = np.ones(nx)
u[int(0.5/dx):int(1/dx)+1] = 2
print(u)

xdimension = np.linspace(0,2,nx)
pyplot.plot(xdimension, u)
```

```python
utemp = np.ones(nx)
for n in range(1, nt): #we will run this loop over nt time steps 
    utemp = u.copy() #make a copy of u
    for i in range(1, nx-1): #we will run a loop over nx points in space, starting on element one
        u[i] = utemp[i] + nu*(dt/(dx**2))*(utemp[i+1]+utemp[i-1]-(2*utemp[i]))
```

```python
pyplot.plot(xdimension, u)
```
